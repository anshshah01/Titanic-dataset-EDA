Performed in-depth exploratory data analysis on Titanic dataset, handling missing data, visualizing distributions and feature relationships using Seaborn and Matplotlib to uncover insights into survival patterns.

#### Results & Insights

Sex: Females had a higher survival rate.

Pclass: 1st class passengers had better survival chances.

Age: children had relatively better survival outcomes.

Embarked: Most passengers boarded from port 'S'.



```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("yasserh/titanic-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df=pd.read_csv('NewTitanic.csv')
```

```python
df['Survived'].value_counts()# so total 152 were survived
```

```python
df.isnull().sum()
```

```python
df['Embarked'].value_counts()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Fare']=df['Fare'].fillna(df['Fare'].median())

df['Embarked']=df['Embarked'].fillna('S')
```

```python
df=df.drop(columns=['Cabin'])
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Number of Male and Female Survivors")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

70/100 female were survived

19/100 male survived
